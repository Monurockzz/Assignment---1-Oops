```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount. Please enter a positive value.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.balance

# Example usage:
account1 = BankAccount("123456", "John Doe", 1000.0)
print(f"Account Holder: {account1.account_holder_name}, Balance: ${account1.get_balance()}")

account1.deposit(500.0)
account1.withdraw(200.0)
```

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_bonus(self):
        return self.salary * 0.1  # 10% of the salary

    def display_details(self):
        print(f"Employee ID: {self.employee_id}, Name: {self.name}, Salary: ${self.salary}")


# Example usage:
employee1 = Employee("E001", "Alice", 50000.0)
bonus = employee1.calculate_bonus()
employee1.display_details()
print(f"Yearly Bonus: ${bonus}")
```

```python
class VehicleRental:
    def __init__(self):
        self.available_vehicles = set()

    def rent_vehicle(self, vehicle):
        self.available_vehicles.discard(vehicle)
        print(f"Rented {vehicle}")

    def return_vehicle(self, vehicle):
        self.available_vehicles.add(vehicle)
        print(f"Returned {vehicle}")

    def display_available_vehicles(self):
        print("Available Vehicles:", self.available_vehicles)


# Example usage:
rental_system = VehicleRental()
rental_system.available_vehicles = {"Car1", "Car2", "Bike1"}
rental_system.display_available_vehicles()
rental_system.rent_vehicle("Car1")
rental_system.display_available_vehicles()
rental_system.return_vehicle("Car1")
rental_system.display_available_vehicles()
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        return f"{self.title} by {self.author}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added book: {book}")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.is_available:
                book.is_available = False
                print(f"Borrowed book: {book}")
                return
        print(f"Book {title} is not available.")

    def display_available_books(self):
        available_books = [book for book in self.books if book.is_available]
        if available_books:
            print("Available Books:")
            for book in available_books:
                print(f"  {book}")
        else:
            print("No available books in the library.")

# Example usage:
library = Library()

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("1984", "George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()

library.borrow_book("To Kill a Mockingbird")
library.display_available_books()

library.borrow_book("To Kill a Mockingbird")  # Trying to borrow the same book again

library.display_available_books()
```

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"{self.name} (ID: {self.product_id}), Price: ${self.price}, Quantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"Added product to inventory: {product}")

    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Updated quantity for {product}: {new_quantity}")
                return
        print(f"Product with ID {product_id} not found in inventory.")

    def display_available_products(self):
        if self.products:
            print("Available Products:")
            for product in self.products:
                print(f"  {product}")
        else:
            print("No products available in the inventory.")

# Example usage:
inventory_system = Inventory()

product1 = Product("P001", "Laptop", 1200.0, 10)
product2 = Product("P002", "Smartphone", 800.0, 15)
product3 = Product("P003", "Headphones", 100.0, 25)

inventory_system.add_product(product1)
inventory_system.add_product(product2)
inventory_system.add_product(product3)

inventory_system.display_available_products()

inventory_system.update_quantity("P002", 20)
inventory_system.display_available_products()

inventory_system.update_quantity("P004", 30)  # Trying to update quantity for a non-existent product

inventory_system.display_available_products()
```

```python
class Student:
    def __init__(self, student_id, name, grades=[]):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def add_grade(self, grade):
        self.grades.append(grade)
        print(f"Added grade {grade} for {self.name}")

    def calculate_average_grade(self):
        if not self.grades:
            return 0.0
        return sum(self.grades) / len(self.grades)

    def display_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {', '.join(map(str, self.grades))}")
        average_grade = self.calculate_average_grade()
        print(f"Average Grade: {average_grade:.2f}")

# Example usage:
student1 = Student("S001", "Alice", [85, 90, 78, 92])
student1.display_details()

# Add more grades for the student
student1.add_grade(88)
student1.add_grade(95)

student1.display_details()
```

```python
class SocialMediaProfile:
    def __init__(self, username, posts=[]):
        self.username = username
        self.posts = posts

    def add_post(self, post):
        self.posts.append(post)
        print(f"New post added by {self.username}")

    def display_posts(self):
        if not self.posts:
            print(f"No posts available for {self.username}")
        else:
            print(f"Posts by {self.username}:")
            for post in self.posts:
                print(f"  {post}")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            print(f"Matching posts for '{keyword}' by {self.username}:")
            for post in matching_posts:
                print(f"  {post}")
        else:
            print(f"No matching posts for '{keyword}' by {self.username}")

# Example usage:
profile1 = SocialMediaProfile("Alice")
profile1.add_post("Enjoying a sunny day! ☀️ #SummerVibes")
profile1.add_post("Just finished reading a great book. Highly recommend!")

profile1.display_posts()

# Add more posts for the profile
profile1.add_post("Trying out a new recipe for dinner. 🍲 #Foodie")
profile1.add_post("Excited for the upcoming vacation! ✈️ #Travel")

profile1.display_posts()

# Search for posts with a specific keyword
profile1.search_posts("book")
```

```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})
        print(f"Added task: {task}, Due Date: {due_date}")

    def mark_task_completed(self, task):
        for t in self.tasks:
            if t["task"] == task:
                t["completed"] = True
                print(f"Marked task '{task}' as completed.")
                return
        print(f"Task '{task}' not found in the ToDo list.")

    def display_pending_tasks(self):
        pending_tasks = [t for t in self.tasks if not t["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for t in pending_tasks:
                print(f"  Task: {t['task']}, Due Date: {t['due_date']}")
        else:
            print("No pending tasks in the ToDo list.")

# Example usage:
todo_list = ToDoList()

todo_list.add_task("Write report", "2024-01-10")
todo_list.add_task("Prepare presentation", "2024-01-15")
todo_list.add_task("Exercise", "2024-01-12")

todo_list.display_pending_tasks()

# Mark a task as completed
todo_list.mark_task_completed("Prepare presentation")

todo_list.display_pending_tasks()
```
